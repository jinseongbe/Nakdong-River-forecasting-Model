```python
import pandas as pd
import numpy as np
```

```python
Up = pd.read_excel('Input(DG)_data.xlsx', index = False)
```

```python
Up_B = Up.filter(like='Temp',axis=1)
Up_C = Up.filter(like='Date',axis=1)
```

```python
# Up.columns[28]
print(list(Up.iloc[:,[0,1,8,28,37,46,55,64,73]].columns))
```

```python
Up_A = Up.drop(Up.iloc[:,[0,1,8,28,37,46,55,64,73]].columns,axis=1)
```

```python
Up_A.info()
```

```python
list(Up_A.min())
```

```python
n = 0

for column in Up_A:
    Data = Up_A.iloc[:,n]
    Data = list(Data)
    while min(Data) == 0:
        Data.remove(0)
    none_zero_min = min(Data)/2
    
    for m in range(0,len(Data)):
        if Up_A.iloc[m,n] == 0:
            Up_A.iloc[m,n] = none_zero_min
            
    n += 1
print('\n all done\n')
```

```python
list(Up_A.min())
```

```python
Up_A.head()
```

```python
Up_A = Up_A.apply(lambda x: np.log(x))
```

```python
result_UP = pd.concat([Up_C, Up_A, Up_B], axis=1, ignore_index=False)
```

```python
result_UP.to_excel('Input(DG)_log.xlsx', index = False)
```
